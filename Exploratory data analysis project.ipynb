### Cleaning the data

```python
# only Want USA races, 50km,or 50miles,2020
# Step 1 : Show 50 miles and 50k
data[data['Event distance/length'] == '50km']
```

```python
# combining the 50km and 50mi to make it one dataframe
data[data['Event distance/length'].isin(['50km','50mi'])]
```

```python
# after combining the 50km and 50mi the next condition is year 2020
data[(data['Event distance/length'].isin(['50km','50mi']))&(data['Year of event'] == 2020)]
```

```python
# step 2 : now we will only take the events happened in USA
data[data['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']
```

```python
data[data['Event name'] .str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
# cobining all the filter together to get a single output
data[(data['Event distance/length'].isin(['50km','50mi']))&(data['Year of event'] == 2020) & (data['Event name'] .str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
# now we have a new data that we have created now named as data1
data1 = data[(data['Event distance/length'].isin(['50km','50mi']))&(data['Year of event'] == 2020) & (data['Event name'] .str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
data1.head()
```

```python
data1.shape
```

```python
shape = (26090, 13)
print(f"rows = {shape[0]}, col = {shape[1]}")
```

```python
data1['Event name'] = data1['Event name'].str.split('(').str.get(0)
```

```python
data1.head(10)
```

```python
# cleaning up the athlete age
data1['athlete_age'] = 2020 - data1['Athlete year of birth']
```

```python
# removing "h" from the athlete performance 
data1['Athlete performance'] = data1['Athlete performance'].str.split(' ').str.get(0)
```

```python
data1.head()
```

```python
# dropping column which is not important for the analysis
# Like we can drop the athlete club,athlete country (as we are only considering the USA country), 
# athlete age category, athlete year of birth
data1= data1.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
data1.head()
```

```python
# Cleaning the null values in the dataset if any
data1.isna().sum()         # this will give us the total count of the null values in each columns in dataset
```

```python
data1[data1['athlete_age'].isna()== 1]  # this will give us the data against the null values for all the columns  
```

```python
# dropping the null values rows from the dataset
data1 = data1.dropna()
```

```python
data1.shape
```

```python
# Checking for the duplicate values in the dataset
data1[data1.duplicated() == True]             # This will check the duplicate values in the dataset
```

```python
# Reset index after dropping the null and duplicate values
data1.reset_index(drop = True)
```

```python
#fixing the datatypes for better efficiency
data1.dtypes
```

```python
data1['athlete_age'] = data1['athlete_age'].astype(int)
data1['Athlete average speed'] = data1['Athlete average speed'].astype(float)
data1['Event name'] = data1['Event name'].astype(str)
```

```python
# Rename the column do that everyone can understand what type of dataset it is.
data1 = data1.rename(columns = {'Year of event':'Year',
                                'Event dates':'Date',
                                'Event name':'Race name',
                                'Event distance/length':'Race_length',
                                'Event number of finishers':'Race no of finishers',
                                'Athlete performance':'Athlete performance',
                                'Athlete gender':'Athlete gender',
                                'Athlete average speed':'Athlete average speed',
                                'Athlete ID':'Athlete ID',
                                'athlete_age':'Athlete age'})
```

```python
data1.head(2)
```

```python
# re-order the columns in correct sequence
data2 = data1[['Race name','Race_length','Date','Year','Race no of finishers','Athlete ID','Athlete gender','Athlete age','Athlete average speed','Athlete performance']]
```

```python
data2.head(2)
```

#### Insights from the datasets

```python
# Difference in speed for the 50km,50mi male to female or gender-wise
data2.groupby(['Race_length','Athlete gender'])['Athlete average speed'].mean()
```

```python
# Age groups are the best in the 50km race (20 + races min)
data2.query('Race_length == "50mi"').groupby('Athlete age')['Athlete average speed'].agg(['mean','count']).sort_values('mean',ascending=False).query('count>19')
```

```python
# # Season for the data --> slower in summer than winters?

# sprint 3-5
# summer 6-8
# fall 9-11
# winter 12-2
# split between two decimals
```

```python
data2['Month'] = data2['Date'].str.split('.').str.get(1).astype(int)
```

```python
data2.head(24)
```

```python
data2['Season'] = data2['Month'].apply(
    lambda x: 'Winter' if x > 11 or x <= 2 else
              'Fall' if x > 8 else
              'Summer' if x > 5 else
              'Spring' if x > 2 else 'Winter')
```

```python
data2.head(2)
```

```python
data2.groupby('Season')['Athlete average speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```
